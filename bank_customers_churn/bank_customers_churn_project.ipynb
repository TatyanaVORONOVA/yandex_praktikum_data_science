# Построение модели для прогнозирования ухода клиентов из банка 

## Введение

**Вводные данные проекта:**
В «Бета-Банке» наблюдается отток клиентов. Нужно спрогнозировать, уйдет клиент из банка в ближайшее время или нет.

**Заказчик:** «Бета-Банка».

**Цель проекта:** построение качественной модели для максимально точного прогнозирования оттока клиентов из банка. 

**Дополнительная информация от заказчика:**
- значением метрики F1-меры должна быть не менее 0.59;
- дополнительно измерить AUC-ROC, сравнить её значение с F1-мерой;
- банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

**Данные для анализа и построения модели:** исторические данные о поведении клиентов и расторжении договоров с банком.
Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

**Этапы проекта:**
1. Загрузить данные, изучить общую информацию об имеющихся данных.
2. Подготовить данные.
3. Исследовать баланс классов, обучить модель без учёта дисбаланса.
3. Улучшить качество модели, учитывая дисбаланс классов.
4. Обучить разные модели и найти лучшую.
5. Провести финальное тестирование.

**Описание данных**

Признаки:
- RowNumber — индекс строки в данных
- CustomerId — уникальный идентификатор клиента
- Surname — фамилия
- CreditScore — кредитный рейтинг
- Geography — страна проживания
- Gender — пол
- Age — возраст
- Tenure — сколько лет человек является клиентом банка
- Balance — баланс на счёте
- NumOfProducts — количество продуктов банка, используемых клиентом
- HasCrCard — наличие кредитной карты
- IsActiveMember — активность клиента
- EstimatedSalary — предполагаемая зарплата

Целевой признак:
- Exited — факт ухода клиента

#### Оценка распределения числовых признаков

```python
# создадим функцию для построения гистограммы histplot и boxplot
def graph(data, data_label, num_bins, log_scale=False):
    sns.set()
    f, axes = plt.subplots(1, 2, figsize=(16, 4))
    axes[0].set_title(data_label + ". Плотность распределения", fontsize=18)
    axes[0].set_ylabel('плотность', fontsize=15)
    sns.histplot(data, bins=num_bins,  ax=axes[0], color='seagreen', kde=True)
    axes[1].set_title(data_label + ". Boxplot", fontsize=18)
    sns.boxplot(data=df, x=data_label, ax=axes[1], color='mediumspringgreen')
    plt.show()
```

```python
# построим графики
columns_name = ['credit_score', 'age', 'balance', 'estimated_salary']
for name in columns_name:
    graph(df[name], name, num_bins=30)
```

В признаках credit_score и age есть выбросы. Посмотрим какой процент данных попадает в них.

```python
max_age = (df['age'].quantile(0.75) - df['age'].quantile(0.25))*1.5 + df['age'].quantile(0.75)
min_credit_score = (df['credit_score'].quantile(0.25)-(df['credit_score'].quantile(0.75)-df['credit_score'].quantile(0.25))*1.5)

print('Доля выбросов в признаке age', round((df.query('age > @max_age')['age'].count()/df.shape[0]*100), 2), '%')
print('Доля выбросов в признаке credit_score', round((df.query('credit_score < @min_credit_score')['credit_score']
                                                      .count()/df.shape[0]*100), 3), '%')
```

```python
print('Общая доля выбросов', 
      round((df.query('age > @max_age | credit_score < @min_credit_score')['age'].count()/df.shape[0]*100),2), '%')
```

Доля выбросов не велика, поэтому лучше очистить данные от них.

### Определение приоритетных метрик качества для моделей

Для того, чтобы корректно оценить значимость той или иной метрики качества модели необходимо четко понимать цели компании, в частности банка. В случае с оттоком действующих клиентов важно придерживаться баланса между двумя крайностями:
1. **понести сверх затраты на удержание клиентов** не планирующих менять банк  - у модели будет перевес в пользу recall, соответственно ложноположительные результаты возрастут;
2. **пропустить клиента, который собирается сменить финансового партнера** - у модели будет перевес в сторону precision, то есть будут отобраны только те клиенты, которые максимально точно подходят под критерии нелояльных клиентов, собирающихся уйти из банка.

Так как accuracy не объективна при дисбалансе классов, а у нас именно такая ситуация (примерно 80% на 20%), то воспользуемся дополнительными метриками для оценки качества модели:
- **F1-мера** (позволят оценить гармоническое среднее между точностью (precision) и полнотой (recall)),
- **auc_roc** (ROC-кривая учитывает не только результаты классификации, но и вероятность предсказания всех классов - насколько хорошо разделяются случаи в разных группах).

Отслеживание одновременно f1-меру (обращая внимание на precision и recall) и auc_roc подберем оптимальную модель для прогнозирования оттока. На основе выбранной модели поработаем с дисбалансом.

#### Подготовка функций для подбора параметров

```python
# функция для подсчета f1 меры на обучающей и валидационной выборках (глубина дерева)
def change_graph (features_train, target_train, features_valid, target_valid, model, class_weight):
    f1_valid = []
    f1_train = []
    if model == RandomForestClassifier:
        for parametr in range(1, 25):
            model = RandomForestClassifier(random_state=12345, max_depth=parametr, class_weight=class_weight)
            model.fit(features_train, target_train)
            predictions_train = model.predict(features_train) # получим предсказание
            result_train_f1 = f1_score(target_train, predictions_train) # посчитаем качество модели
            predictions_valid = model.predict(features_valid)
            result_valid_f1 = f1_score(target_valid, predictions_valid) 
            f1_train.append(result_train_f1)
            f1_valid.append(result_valid_f1)
    elif model == DecisionTreeClassifier:
        for parametr in range(1, 25):
            model = DecisionTreeClassifier(random_state=12345, max_depth=parametr, class_weight=class_weight)
            model.fit(features_train, target_train)
            predictions_train = model.predict(features_train) # получим предсказание
            result_train_f1 = f1_score(target_train, predictions_train) # посчитаем качество модели
            predictions_valid = model.predict(features_valid)
            result_valid_f1 = f1_score(target_valid, predictions_valid) 
            f1_train.append(result_train_f1)
            f1_valid.append(result_valid_f1)
        f1_train = pd.Series(f1_train)
        f1_valid = pd.Series(f1_valid)    
    data_metric = {'f1_train':f1_train, 'f1_valid':f1_valid}
    return data_metric 
```

```python
# функция для подсчета f1 меры на обучающей и валидационной выборках (количество деревьев для случайного леса)
def change_graph2 (features_train, target_train, features_valid, target_valid, max_depth, class_weight):
    f1_valid = []
    f1_train = []
    for parametr in range(1, 50):
        model = RandomForestClassifier(random_state=12345, max_depth=max_depth, n_estimators=parametr,class_weight=class_weight)
        model.fit(features_train, target_train)
        predictions_train = model.predict(features_train) # получим предсказание
        result_train_f1 = f1_score(target_train, predictions_train) # посчитаем качество модели
        predictions_valid = model.predict(features_valid)
        result_valid_f1 = f1_score(target_valid, predictions_valid) 
        f1_train.append(result_train_f1)
        f1_valid.append(result_valid_f1)
    
    f1_train = pd.Series(f1_train)
    f1_valid = pd.Series(f1_valid)    
    data_metric = {'f1_train':f1_train, 'f1_valid':f1_valid}
    return data_metric 
```

Для борьбы с дисбалансом классов будем использовать:

### Подготовка выборок для корректировки дисбаланса

```python
# посчитаем соотношение классов
rat_upsampled = round(len(target.loc[target==0])/len(target.loc[target==1]))
rat_upsampled
```

```python
# посчитаем соотношение классов
rat_downsampled = round(len(target.loc[target==1])/len(target.loc[target==0]), 2)
rat_downsampled
```

```python
# создадим функцию для увеличения выборки меньшего класса 1
def upsample(features, target, rat_upsampled):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * rat_upsampled)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * rat_upsampled)
    
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)
    return features_upsampled, target_upsampled
```

```python
# создадим функцию для уменьшения большего класса 0
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
# выборка для upsampling
features_upsampled, target_upsampled = upsample(features_train, target_train, rat_upsampled)
```

```python
# выборка для downsampling
features_downsampled, target_downsampled = downsample(features_train, target_train, rat_downsampled)
```
